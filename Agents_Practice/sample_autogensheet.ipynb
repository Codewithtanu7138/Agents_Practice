HOW TO MAKE MULTI-AGENTIC SYSTEM.


1.   RoundRobin team: Making the agents work sequentially in a team but in a common context i.e. first agent will be passing it's output to the following agents along with the context(Here context means user's promt): RoubdRobin is basically acting as an orchestrator/agents workflow decider.
2.   All with Assistant Agents



```python
storywriter_agent= AssistantAgent(
    name="storywriter_agent",
    model_client=model_client,
    system_message=''' You will write a creative story based on the user's input for a child, so don't include any harmful context
    STRICTLY GIVE STORY TONE, NO OTHER ANSWERS
    '''
)

storyreviewer= AssistantAgent(
    name="storyreviewer",
    model_client=model_client,
    system_message=''' Based on the output from storywriter_agent you need to review the story such that it starts and ends in 7 lines
    and the story shall be strictly for child '''
)
```

```python
from autogen_agentchat.teams import RoundRobinGroupChat
team_work = RoundRobinGroupChat(
    participants=[storywriter_agent, storyreviewer],
    max_turns=2

)
```

```python
response = await team_work.run(task="Write a story for Lord Shiva")
print(response.messages[-1].content)
```

```python
response= await Console(team_work.run_stream(task="Write a story for Lord Shiva"))
print(response.messages[-1].content)
```

//CREATING 4 AGENTS

```python
storywriter_agent= AssistantAgent(
    name="storywriter_agent",
    model_client=model_client,
    system_message=''' You will write a creative story based on the user's input for a child, so don't include any harmful context
    STRICTLY GIVE STORY TONE, NO OTHER ANSWERS
    '''
)

storygrammar= AssistantAgent(
    name="storygrammar",
    model_client=model_client,
    system_message=''' Based on the output from storywriter_agent you need to review the story such that it has correct grammar, words, punctuations, commas and full stops
    and the story shall be strictly for child '''
)

story_line= AssistantAgent(
    name="story_line",
    model_client=model_client,
    system_message=''' You will write a creative story based on the user's input for a child, in way that each lines are coming on a new line
    STRICTLY GIVE STORY TONE, NO OTHER ANSWERS
    '''
)

storyreviewer= AssistantAgent(
    name="storyreviewer",
    model_client=model_client,
    system_message=''' Based on the output from storywriter_agent you need to review the story such that it starts and ends in 7 lines
    and the story shall be strictly for child '''
)
```

```python
from autogen_agentchat.teams import RoundRobinGroupChat
team_work = RoundRobinGroupChat(
    participants=[storywriter_agent, storygrammar, story_line, storyreviewer],
    max_turns=4

)
```

```python
response = await team_work.run(task="Write a story for Lord Shiva")
print(response.messages[-1].content)


```

```python
response2= await Console(team_work.run_stream(task="Write a story for Lord Shiva"))
print(response2.messages[-1].content)
```
